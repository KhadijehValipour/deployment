```python
response = requests.get("https://picsum.photos/200/300")
response.status_code
```

```python
response.content
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

file_data = np.frombuffer(response.content, dtype="int8")
image = cv2.imdecode(file_data, cv2.IMREAD_UNCHANGED)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```
